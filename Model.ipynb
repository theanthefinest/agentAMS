```python
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('','', string.punctuation))
    return text
```

```python
def preprocess_input(course_name, vectorizer):
    course_name = preprocess_text(course_name)
    course_name_lemmatized = lemmatize_text(course_name)
    course_vector = vectorizer.transform([course_name_lemmatized]).toarray()
    return course_vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
import numpy as np

def precision_at_k(predicted_scores, true_relevance, k=10):
    top_k_indices = np.argsort(predicted_scores)[::-1][:k]
    relevant_courses = [1 if true_relevance[i] > 0 else 0 for i in top_k_indices]
    precision = np.sum(relevant_courses) / k
    return precision

def recall_at_k(predicted_scores, true_relevance, k=10):
    top_k_indices = np.argsort(predicted_scores)[::-1][:k]
    relevant_courses = [1 if true_relevance[i] > 0 else 0 for i in top_k_indices]
    recall = np.sum(np.array(relevant_courses)) / np.sum(np.array(true_relevance) > 0)
    return recall

def mean_squared_error_custom(true_relevance, predicted_scores):
    mse = mean_squared_error(true_relevance, predicted_scores)
    return mse

def mean_average_precision(predicted_scores, true_relevance, k=10):
    average_precision = 0
    for i in range(len(true_relevance)):
        top_k_indices = np.argsort(predicted_scores[i])[::-1][:k]
        relevant_courses = [1 if true_relevance[i][idx] > 0 else 0 for idx in top_k_indices]
        precision_at_i = np.cumsum(relevant_courses) / (np.arange(k) + 1)
        num_relevant = np.sum(relevant_courses)
        if num_relevant > 0:
            avg_precision = np.sum(precision_at_i * relevant_courses) / num_relevant
            average_precision += avg_precision
    map_score = average_precision / len(true_relevance)
    return map_score


def evaluate_model(df, predicted_scores, true_relevance, top_k=10):
    precision = precision_at_k(predicted_scores, true_relevance, k=top_k)
    recall = recall_at_k(predicted_scores, true_relevance, k=top_k)
    mse = mean_squared_error_custom(true_relevance, predicted_scores)
    if isinstance(true_relevance[0], list):
        map_score = mean_average_precision(np.array([predicted_scores]), np.array([true_relevance[0]]), k=top_k)
    else:
        map_score = "N/A (single instance)"

    print("===" *20)
    print(f"    Precision@{top_k}: {precision:.4f}")
    print(f"    Recall@{top_k}: {recall:.4f}")
    print(f"    Mean Squared Error (MSE): {mse:.4f}")
    print("===" *20)

def preprocess_input(course_name, vectorizer):
    course_vector = vectorizer.transform([course_name])
    return course_vector

def recommend(course_name, df, vectorizer, vectors, top_n=10, rating_weight=0.9):

    course_vector = preprocess_input(course_name, vectorizer)

    similarity_scores = cosine_similarity(course_vector, vectors)

    similarity_scores = similarity_scores.flatten()

    similar_courses = [(index, similarity_scores[index], df.iloc[index]['Course Rating'])
                       for index in range(len(similarity_scores))]

    combined_scores = [(index, similarity_score * (1 - rating_weight) + rating * rating_weight)
                       for index, similarity_score, rating in similar_courses]

    combined_scores_sorted = sorted(combined_scores, key=lambda x: x[1], reverse=True)

    predicted_scores = [combined_scores_sorted[i][1] for i in range(top_n)]
    true_relevance_for_top_n = [df.iloc[combined_scores_sorted[i][0]]['Course Rating'] for i in range(top_n)]


    print("===" * 60)
    print(f"{'Top ' + str(top_n) + ' Recommendations for:':<40} '{course_name}'")
    print("===" * 60)
    print(f"{'Rank':<5} {'Course Name':<40} {'Rating':>8}")
    print("-" * 60)
    for rank, (index, score) in enumerate(combined_scores_sorted[:top_n]):
        course = df.iloc[index]
        print(f"{rank + 1:<5} {course['course_name']:<40} {course['Course Rating']:>8.2f}")
    print("=" * 60)
    evaluate_model(df, predicted_scores , true_relevance_for_top_n, top_k=top_n)


recommend("Introduction to Machine Learning", df, vectorizer, vectors, top_n=5, rating_weight=0.3)
```

```python
def recommend_knn_content(course_name, df, tfidf_matrix, knn_model, top_n=5):
    try:
        course_index = df[df['course_name'] == course_name].index[0]
    except IndexError:
        print(f"Course '{course_name}' not found.")
        return

    course_vector = tfidf_matrix[course_index]
    distances, indices = knn_model.kneighbors(course_vector, n_neighbors=top_n + 1)

    recommended_courses = []
    for i in range(1, len(distances[0])):
        index = indices[0][i]
        recommended_courses.append((df['course_name'][index], df['Course Rating'][index], 1 - distances[0][i]))

    print("===" * 20)
    print(f"{'Top ' + str(top_n) + ' to:':} '{course_name}'")
    print(f"{'Rank':<5} {'Course Name':<50} {'Rating':>8}")
    print("===" * 20)
    for name, rating, similarity in recommended_courses:
        print(f"{name :<50} {rating:>8.2f}")
        #print(f"- {name}          Rating: {rating:.2f}")
        
    print("-" * 60)
```

```python
from sklearn.neighbors import NearestNeighbors
n_neighbors_recommendation = 5
knn_recommend_model = NearestNeighbors(n_neighbors=n_neighbors_recommendation, metric='cosine')
knn_recommend_model.fit(vectors)
```

```python
recommend_knn_content("Introduction to Machine Learning", df, vectors, knn_recommend_model, top_n=10)
```

```python
def evaluate_knn_rating_prediction(X_train, X_test, y_train, y_test, n_neighbors=5):
    knn_rating_model = NearestNeighbors(n_neighbors=n_neighbors, metric='cosine')
    knn_rating_model.fit(X_train)

    predicted_ratings = []
    for i in range(X_test.shape[0]):
        test_vector = X_test[i]
        distances, indices = knn_rating_model.kneighbors(test_vector, n_neighbors=n_neighbors)

        neighbor_ratings = y_train.iloc[indices[0]].values

        predicted_rating = np.mean(neighbor_ratings)
        predicted_ratings.append(predicted_rating)

    mse = mean_squared_error(y_test, predicted_ratings)
    print(f"\n--- KNN Rating Prediction Evaluation (n_neighbors={n_neighbors}) ---")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
```

```python
evaluate_knn_rating_prediction(X_train, X_test, y_train, y_test, n_neighbors=3)
evaluate_knn_rating_prediction(X_train, X_test, y_train, y_test, n_neighbors=5)
evaluate_knn_rating_prediction(X_train, X_test, y_train, y_test, n_neighbors=7)
```
